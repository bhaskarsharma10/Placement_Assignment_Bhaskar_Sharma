```python
import pandas as pd
```

```python
df = pd.read_json('https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json')
```

```python
df = pd.json_normalize(df['pokemon']) # normalizing the json data to readable columns
```

```python
df.head()
```

```python
df_backup=df
df.info()
```

```python
df['candy_count']=df['candy_count'].fillna(0).astype(int)
```

```python
df['num']=df['num'].astype(int)
df_backup=df
```

```python
df['height']=df['height'].str.extract('(\d+\.\d+)').astype(float)
```

```python
df['weight']=df['weight'].str.extract('(\d+\.\d+)').astype(float)
df_backup=df
```

```python
df['egg']=df['egg'].str.extract('(\d+)').fillna(0).astype(float)
```

```python
df['avg_spawns']=df['avg_spawns'].astype(int)
```

```python
df['spawn_time']=pd.to_timedelta(df['spawn_time']+':00', errors='coerce') #converter hh:mm to hh:mm:ss format and not time values to Nat
```
